```python
from langgraph.graph import StateGraph, START, END
from typing import TypedDict
```

```python
class Cricketer(TypedDict):
    
    total_runs: int
    balls_faced: int
    total_boundaries: int
    total_outs: int
    strike_rate: float
    boundary_percentage: float
    average: float
    summary: str
```

```python
def strike_rate(state: Cricketer):

    total_runs = state['total_runs']
    balls_faced = state['balls_faced']

    strike_rate = (total_runs / balls_faced) * 100

    return {
        'strike_rate': strike_rate
    }

def boundary_percentage(state: Cricketer) -> Cricketer:

    total_runs = state['total_runs']
    total_boundaries = state['total_boundaries']

    boundary_percentage = (total_boundaries / total_runs) * 100

    return {
        'boundary_percentage': boundary_percentage
    }


def average_runs(state: Cricketer) -> Cricketer:

    total_runs = state['total_runs']
    total_outs = state['total_outs']

    average_runs = total_runs / total_outs

    return {
        'average': average_runs
    }

def show_answer(state: Cricketer) -> Cricketer:

    print("Strike Rate :", state['strike_rate'])
    print("Boundary Percentage :", state['boundary_percentage'])
    print("Average :", state['average'])

    return state
```

```python
# define your graph
graph = StateGraph(Cricketer)

# add nodes
graph.add_node('strike_rate', strike_rate)
graph.add_node('boundary_percentage', boundary_percentage)
graph.add_node('average_runs', average_runs)
graph.add_node('show_answer', show_answer)


# add edges
graph.add_edge(START, 'strike_rate')
graph.add_edge(START, 'boundary_percentage')
graph.add_edge(START, 'average_runs')
graph.add_edge('strike_rate', 'show_answer')
graph.add_edge('boundary_percentage', 'show_answer')
graph.add_edge('average_runs', 'show_answer')
graph.add_edge('show_answer', END)

# compile the graph
workflow = graph.compile()
```

```python
initial_state = {
    'total_runs': 17266,
    'balls_faced': 19218,
    'total_boundaries': 1578,
    'total_outs': 452
}

result = workflow.invoke(initial_state)

print(result)
```
