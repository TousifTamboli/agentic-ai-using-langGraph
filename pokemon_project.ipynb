```python
from langchain_openai import ChatOpenAI
import requests
```

```python
model = ChatOpenAI(
    model="gpt-5.4-mini",
    temperature=0.7
)
```

```python
while True:
    print("Enter exit to exit")
    pokemon_name = input("Enter name").strip().lower()

    if(pokemon_name == "exit"):
        break

    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name}"

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        
        ai_response = model.invoke(
            f"Give me a detailed summary of this Pokémon information:\n{data}"
        )
        
        print(ai_response.content)
    else:
        print("Pokemon not found")
```
